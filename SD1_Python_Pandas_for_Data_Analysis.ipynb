# What is Pandas?

"Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language" (Pandas https://pandas.pydata.org/). In practice Pandas is a Python library that makes data analysis in Python both bearable (a matter of opinion) and powerful.

Because Python is open source, it has many libraries developed by the community for different purposes. Pandas has emerged as the standard library for data manipulation and analysis. Often times, a search provides many suggestions for doing a certain data task in Python. While base Python has data capabilities, Pandas is the preferred library for data work.

##Univariate regression

```python
# Drop missing values from the columns used in the regression
mydata_clean = data.dropna(subset=['HRLYEARN', 'TENURE'])

# Create the model
model = smf.ols('HRLYEARN ~ TENURE', data=mydata_clean)

# Fit the model
results = model.fit()

# Display regression results
print(results.summary())
```
